## Step 1.1 : Load the data file into a data frame, setting the Rank column as the index

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
vgsales = pd.read_csv('vgsales.csv', index_col = 'Rank')
```

```python
vgsales['Name'].head(10)
```

## Step 1.3 : Report the total number of titles

```python
print("The total number of titles = {}" .format(vgsales['Name'].count()))
```

## Step 1.4 : Report some summary statistics about the distribution of global sales

```python
print("The mean is : {}" .format(np.mean(vgsales['Global_Sales'])))
print("The median is : {}" .format(np.median(vgsales['Global_Sales'])))
print("The standard deviation is : {}" .format(np.std(vgsales['Global_Sales'])))
```

## Step 1.5 : What proportion (or percentage) of titles earned at least 1 million dollars in global sales?

```python
print("The  proportion of titles that earned at least $1 million in global sales : {}" .format((vgsales['Global_Sales'][vgsales['Global_Sales'] >= 1].count())/vgsales['Name'].count()))
```

# EXERCISE #2

## Step 2.1 : Filter the DataFrame to show  the top 3 sellers in the Action genre

```python
print ('The top 3 sellers in the Action genre are:')
print (vgsales[vgsales['Genre'] == 'Action'].sort_values(by = 'Global_Sales',ascending = False)['Name'].head(n=3))
```

## Step 2.2 : Generalize the filter you created above and write a function top_sellers for the same

```python
def top_sellers (vgsales,cat=None,val=None,n=5):
    if cat != None and val != None:
        df_top = vgsales[vgsales[cat] == val].head(n)
    else:
        df_top = vgsales.head(n)
    return df_top
```

## Step 2.3: Use the function to display the overall top 5 sellers

```python
top_sellers(vgsales)
```

## Use the function to display the top 10 NES games

```python
top_sellers(vgsales,'Platform','NES',10)
```

## Use the function to display top 10 selling games in the year 2000

```python
top_sellers(vgsales,'Year',2000,10)
```

## Use the function to determine top 10 selling sports games

```python
top_sellers(vgsales,'Genre','Sports',10)
```

## Step 3.1 : Generate a (linear) correlation matrix between sales in specific markets (North America, Europe, Japan, Other) and global sales.

```python
sales = vgsales[vgsales.columns[-6:]]
sales.corr()
```

## Relation of sales between various markets is positively correlated.

## Step 3.2 : Repeat this calculation for the subset of video games released in the year 2000, and compare these results to those for the subset of video games released in the year 2015. What do you observe about how the relationship between the sales figures has changed over the 15-year period?

```python
data = pd.read_csv("vgsales.csv")
year_2000=data.loc[data['Year'] == 2000.0]
year_2000_sales = year_2000[year_2000.columns[-6:]]
print("Correlation matrix in the year 2000 : ")
print()
print(year_2000_sales.corr())
print()
print("Correlation matrix in the year 2015 : ")
print()
year_2015=data.loc[data['Year'] == 2015.0]
year_2015_sales = year_2015[year_2015.columns[-6:]]
print(year_2015_sales.corr())
```

## The global sales correlation coefficient for the year 2015 in NA was higher than it was in 2000
## The global sales correlation coefficient for the year 2015 in EU was higher than it was in 2000
## The global sales correlation coefficient for the year 2015 in JP was lower than it was in 2000
## The global sales correlation coefficient for the year 2015 in Other regions was higher in than it was in 2000
